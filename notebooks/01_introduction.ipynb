# What is machine learning? (20 min, or 8/12 etc)

- Application of algorithms that learn from examples
- Exists at the intersection of statistics, computer engineering, and artificial intelligence
- Representation and generalization
- Useful in every day life
    - Email spam, handwriting analysis, stock market analysis, Netflix
- Especially useful in data analysis
    - Feature extraction, linear regression, classification, clustering
- Vocab
    - Instance/observation/data point
    - Feature
    - Class
    - Categorical
    - Nominal
    - Ordinal
    - Continuous
    - Overfitting **super important**
- Supervised learning
    - Labeled data
    - You know what you’re looking for
    - Classification: predict categorical labels	
    - Regression: predict continuous target variables
- Unsupervised learning
    - Unlabeled data
    - You might have no idea what you’re looking for
    - Clustering: splitting observations into groups
    - Dimensionality reduction: flatten data to fewer dimensions
- Types of machine learning
    - Supervised
    - Unsupervised
    - Semi-supervised
    - Reinforcement learning
    - Neural networks
