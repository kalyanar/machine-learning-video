# Example of linear regression

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split

boston = load_boston()
lr = LinearRegression()

x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target)

lr.fit(x_train, y_train)
lr.predict(x_test)
```

# Example of ridge regression

```python
from sklearn.linear_model import Ridge

lr_ridge = Ridge()
lr_ridge.fit(x_train, y_train)
lr_ridge.predict(x_test)
```

# Example of lasso regression

```python
from sklearn.linear_model import Lasso

lr_lasso = Lasso()
lr_lasso.fit(x_train, y_train)
lr_lasso.predict(x_test)
```

# Comparing models

```python
from sklearn.cross_validation import cross_val_score

cross_val_score(lr, boston.data, boston.target)
```

```python
cross_val_score(lr_ridge, boston.data, boston.target)
```

```python
cross_val_score(lr_lasso, boston.data, boston.target)
```
