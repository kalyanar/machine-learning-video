```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

digits = load_digits()
pca = PCA()

pca_fit = pca.fit(digits.data)
pca_fit.components_
```

```python
pca_fit.transform(digits.data)
```

```python
pca_tuned = PCA(n_components=30)
tuned = pca_tuned.fit(digits.data)
tuned.components_
```

```python
pca_tuned.transform(digits.data)
```

```python
pca.explained_variance_ratio_
```

```python
pca_tuned.explained_variance_ratio_
```
